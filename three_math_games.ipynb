# Scatter Plot Game

*   Randomly generate points on a graph and the player has to input the (x,y) coordinates
*   For added difficulty, make the graph larger

```python
import matplotlib.pyplot as plt
import random

score = 0
guess_limit = 3

levels = [
    {'xmin': -10, 'xmax': 10, 'ymin': -10, 'ymax': 10},
    {'xmin': -20, 'xmax': 20, 'ymin': -20, 'ymax': 20},
    {'xmin': -30, 'xmax': 30, 'ymin': -30, 'ymax': 30}
]

fig, ax = plt.subplots()

def generate_random_point(level):
    level_info = levels[level]
    xmin, xmax, ymin, ymax = level_info['xmin'], level_info['xmax'], level_info['ymin'], level_info['ymax']
    xpoint = random.randint(xmin, xmax)
    ypoint = random.randint(ymin, ymax)
    return xpoint, ypoint

def plot_point(point, level, guessed=False, correct=False):
    level_info = levels[level]
    xmin, xmax, ymin, ymax = level_info['xmin'], level_info['xmax'], level_info['ymin'], level_info['ymax']
    plt.clf()
    plt.axis([xmin, xmax, ymin, ymax])  # window size
    plt.plot([xmin, xmax], [0, 0], 'b')  # blue x axis
    plt.plot([0, 0], [ymin, ymax], 'b')  # blue y axis

    if guessed:
        marker = 's' if correct else 'x'
        color = 'green' if correct else 'red'
        alpha = 1.0 if correct else 0.7
        plt.scatter(*point, marker=marker, color=color, s=100, alpha=alpha)
    else:
        plt.scatter(*point, marker='o', color='red', s=100)

    plt.grid()  # displays grid lines on graph
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Scatter Plot Game (Level {})'.format(level + 1))
    plt.show()

def validate_input(input_str):
    while True:
        try:
            x, y = map(int, input(input_str).split(","))
            if xmin <= x <= xmax and ymin <= y <= ymax:
                return x, y
            else:
                print("Please enter coordinates within the specified range.")
        except ValueError:
            print("Invalid input. Please enter two integers separated by a comma.")

def play_game(level):
    global score
    point = generate_random_point(level)
    plot_point(point, level, guessed=False, correct=False)
    print()
    guess_count = 0
    while guess_count < guess_limit:
        guess_count += 1
        guess = validate_input("Enter the coordinates of the red point (x,y): ")
        if guess == point:
            score += 1
            print("Correct!")
            plot_point(point, level, guessed=True, correct=True)
            break
        else:
            print("Incorrect!")
            plot_point(point, level, guessed=True, correct=False)
    else:
        print("Out of guesses. Moving on to the next point.")

# Play the game for each level
for level in range(len(levels)):
    print("Level", level + 1)
    play_game(level)

# Show final score
print("Your score:", score)
```

# Algebra Practice Game


*   Generate one-step and two-step problems with random integer values and the player has to input the answer

*   Use positive and negative values. For added difficulty, make the numbers larger


```python
import random

# Function to generate a random integer within a given range
def generate_random_integer(min_value, max_value):
    return random.randint(min_value, max_value)

# Function to generate a one-step equation problem
def generate_one_step_problem():
    a = generate_random_integer(-100, 100)
    b = generate_random_integer(-100, 100)
    c = generate_random_integer(-100, 100)
    problem = f"{a}x + {b} = {c}"
    answer = (c - b) / a
    return problem, answer

# Function to generate a two-step equation problem
def generate_two_step_problem():
    a = generate_random_integer(-100, 100)
    b = generate_random_integer(-100, 100)
    c = generate_random_integer(-100, 100)
    d = generate_random_integer(-100, 100)
    problem = f"{a}x + {b} = {c} + {d}x"
    answer = (c - b) / (a - d)
    return problem, answer

# Function to validate the player's answer
def validate_answer(player_answer, correct_answer):
    return player_answer == correct_answer

# Main game loop
def algebra_practice_game():
    print("Welcome to the Algebra Practice Game!")
    print("Level: One-Step Equations")
    print("Instructions: Solve the following one-step equations.")

    for _ in range(2):
        problem, answer = generate_one_step_problem()
        print("Problem:", problem)
        try:
            player_answer = float(input("Your answer: "))
            if validate_answer(player_answer, answer):
                print("Correct!")
            else:
                print(f"Incorrect! The correct answer is {answer}.")
        except ValueError:
            print("Invalid input! Please enter a numeric value.")

    print("\nLevel: Two-Step Equations")
    print("Instructions: Solve the following two-step equations.")

    for _ in range(2):
        problem, answer = generate_two_step_problem()
        print("Problem:", problem)
        try:
            player_answer = float(input("Your answer: "))
            if validate_answer(player_answer, answer):
                print("Correct!")
            else:
                print(f"Incorrect! The correct answer is {answer}.")
        except ValueError:
            print("Invalid input! Please enter a numeric value.")

# Start the game
algebra_practice_game()
```

# Projectile Game


:*    Display a "wall" with random height and location. Player has to move sliders to adjust a parabolic path to clear the wall

*   For added difficulty, make a second level where players enter a, b, and c without sliders


```python
import random
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, FloatSlider, Button, HBox, VBox
from IPython.display import display
%matplotlib inline

# Define the physics constants
g = 9.8  # Acceleration due to gravity (m/s^2)
x0 = 0  # Initial x-position
y0 = 0  # Initial y-position

# Randomize the wall's height and location
wall_height = random.uniform(5, 15)  # Random wall height between 5 and 15 meters
wall_x = random.uniform(20, 30)  # Random wall x-coordinate between 20 and 30 meters

# Print the height of the wall
print(f"The randomly positioned wall has a height of {wall_height:.2f} meters.")

# Define the projectile motion function for the first level
def projectile_motion(v0, angle):
    angle_rad = np.radians(angle)
    # Calculate time of flight
    t_flight = 2 * v0 * np.sin(angle_rad) / g

    # Calculate time intervals for evaluating position
    t = np.linspace(0, t_flight, num=100, endpoint=True)

    # Calculate x and y coordinates at each time interval
    x = x0 + v0 * np.cos(angle_rad) * t
    y = y0 + v0 * np.sin(angle_rad) * t - 0.5 * g * t ** 2

    return x, y

# Define the projectile motion function for the second level
def projectile_motion_2nd_level(a, b, c):
    # Calculate x and y coordinates
    x = np.linspace(0, 30, num=100)  # Assuming a range of 0-30 meters
    y = a * x ** 2 + b * x + c

    return x, y

# Define the graphing function for the first level
def plot_projectile(v0, angle):
    x, y = projectile_motion(v0, angle)

    # Plot the wall
    fig, ax = plt.subplots()
    ax.plot([wall_x, wall_x], [0, wall_height], 'r--')
    ax.set_xlim(0, wall_x + 5)  # Adjusted to include the wall position
    ax.set_ylim(0, max(20, wall_height + 5))  # Adjusted to accommodate the wall and projectile

    ax.set_xlabel('Distance (m)')
    ax.set_ylabel('Height (m)')

    # Plot the graph
    ax.plot(x, y)
    ax.scatter(x[-1], y[-1], color='red')

    # Check if the rocket clears the wall
    if max(y) > wall_height:
        print("Congratulations! The rocket cleared the wall.")
        level_box.children = [level_2_box]  # Progress to Level 2
    else:
        print("Oops! The rocket did not clear the wall.")

# Define the graphing function for the second level
def plot_projectile_2nd_level(a, b, c):
    x, y = projectile_motion_2nd_level(a, b, c)

    # Plot the wall
    fig, ax = plt.subplots()
    ax.plot([wall_x, wall_x], [0, wall_height], 'r--')
    ax.set_xlim(0, wall_x + 5)  # Adjusted to include the wall position
    ax.set_ylim(0, max(20, wall_height + 5))  # Adjusted to accommodate the wall and projectile

    ax.set_xlabel('Distance (m)')
    ax.set_ylabel('Height (m)')

    # Plot the graph
    ax.plot(x, y)
    ax.scatter(x[-1], y[-1], color='red')

    # Check if the rocket clears the wall
    if max(y) > wall_height:
        print("Congratulations! The rocket cleared the wall.")
    else:
        print("Oops! The rocket did not clear the wall.")

# Set up the sliders for the first level
v0_slider = FloatSlider(value=0, min=10, max=30, step=1, description='Velocity (m/s)')
angle_slider = FloatSlider(value=0, min=0, max=90, step=1, description='Angle (degrees)')
launch_button = Button(description='Launch')

def launch_level1(b):
    plot_projectile(v0_slider.value, angle_slider.value)
    level_1_box.children[-1].disabled = True

launch_button.on_click(launch_level1)

level_1_box = VBox([v0_slider, angle_slider, launch_button])

# Prompt for the coefficients a, b, and c for the second level
a_text = FloatSlider(value=0, min=-10, max=10, step=0.1, description='Coefficient a:')
b_text = FloatSlider(value=0, min=-10, max=10, step=0.1, description='Coefficient b:')
c_text = FloatSlider(value=0, min=-10, max=10, step=0.1, description='Coefficient c:')
launch_button_2 = Button(description='Launch')

def launch_level2(b):
    plot_projectile_2nd_level(a_text.value, b_text.value, c_text.value)
    level_2_box.children[-1].disabled = True

launch_button_2.on_click(launch_level2)

level_2_box = VBox([a_text, b_text, c_text, launch_button_2])
level_2_box.layout.visibility = 'hidden'

# Combine both levels in a single box
level_box = VBox([level_1_box, level_2_box])

display(level_box)

def update_visibility():
    if level_box.children[0] == level_1_box:
        level_2_box.layout.visibility = 'hidden'
    else:
        level_2_box.layout.visibility = 'visible'

level_box.observe(lambda change: update_visibility(), 'children')
```
